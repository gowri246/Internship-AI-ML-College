```python
import pandas as pd
pf=pd.read_csv("/content/preprocessing_practice_dataset.csv")
```

```python
pf.head()
```

```python
pf.isnull().sum()
```

```python
pf['Age_mean_imputed']=pf['Age'].fillna(pf['Age'].mean())
pf['Department_mode_imputed']=pf['Department'].fillna(pf['Department'].mode()[0])
pf['Monthly_Salary_mean_imputed']=pf['Monthly_Salary'].fillna(pf['Monthly_Salary'].mean())
pf['Education_mode_imputed']=pf['Education'].fillna(pf['Education'].mode()[0])
pf['City_mode_imputed']=pf['City'].fillna(pf['City'].mode()[0])
```

```python
pf
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10,6))
sns.barplot(data=pf,x="Age",y="Gender",palette='pastel')
plt.title("Age vs Gender")
plt.xlabel("Age")
plt.ylabel("Gender")
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10,6))

sns.boxplot(data=pf,x="Department",y="Experience_Years",palette='pastel')
plt.title("Department vs Experience")
plt.xlabel("Department")
plt.ylabel("Experience")
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10,6))
sns.lineplot(data=pf,x="City",y="Gender",palette='pastel')
plt.title("City vs Gender")
plt.xlabel("City")
plt.ylabel("Gender")
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(10,6))
sns.countplot(data=pf,x="Department",palette='pastel')
plt.title("Department Distribution")
plt.xlabel("Department")
plt.ylabel("Count")
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(10,6))
sns.histplot(data=pf,x="Monthly_Salary",palette='pastel')
plt.title("Monthly Salary Distribution")
plt.xlabel("Monthly Salary")
plt.ylabel("Frequency")
```

```python
from sklearn.preprocessing import LabelEncoder
encoded_data=pf.copy()
one_hot_encoded=pd.get_dummies(encoded_data['Department'],prefix="Department")
encoded_data=pd.concat([encoded_data,one_hot_encoded],axis=1)
encoded_data
```

```python
performance_class={1:"one",2:"two",3:"three",4:"four",5:"five"}
encoded_data['Performance_Class']=encoded_data['Performance_Rating'].map(performance_class)
encoded_data
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
min_max_scaler=MinMaxScaler()
standard_scaler=StandardScaler()
encoded_data['Min_Max_Age']=min_max_scaler.fit_transform(encoded_data[['Age']])
encoded_data['standard_Salary']=standard_scaler.fit_transform(encoded_data[['Monthly_Salary']])
encoded_data
```
