```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
encoded_data=tf.copy()
label_encoder= LabelEncoder()
encoded_data['Sex_Label'] = label_encoder.fit_transform(encoded_data['Sex'])
```

```python
one_hot_encoded_data = pd.get_dummies(encoded_data['Embarked'],prefix='Embarked')
encoded_data = pd.concat([encoded_data,one_hot_encoded_data],axis=1)
```

```python
pclass_mapping = {1: 'First', 2: 'Second', 3: 'Third'}
```

```python
encoded_data['Pclass_Category'] = encoded_data['Pclass'].map(pclass_mapping)
```

```python
print(encoded_data[['Sex', 'Sex_Label', 'Embarked', 'Embarked_Q', 'Embarked_S', 'Embarked_C', 'Pclass', 'Pclass_Category']].head())
```
